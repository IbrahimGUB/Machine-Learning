```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged_df = pd.concat([dummies, df], axis='columns')
merged_df
```

```python
final_df= merged_df.drop(['Mercedez Benz C class','Car Model'], axis='columns')
final_df
```

```python
%matplotlib inline
plt.scatter(final_df[['Audi A5'],['BMW X5'], ['Mileage'], ['Age(yrs)']],df['Sell Price($)'], color='green', markers="+")
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
reg.fit(final_df[['Audi A5','BMW X5', 'Mileage', 'Age(yrs)']],df['Sell Price($)'])
```

```python
reg.predict([[0,0,45000,4]])
```

```python
reg.predict([[0,1,86000,7]])
```

```python
reg.score(final_df[['Audi A5','BMW X5', 'Mileage', 'Age(yrs)']],df['Sell Price($)'])
```
