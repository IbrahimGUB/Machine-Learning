# K-Means Cluster

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans() #Creating KMeans obj.
```

```python
km.fit(df[['score','income']]) #fit()-> trainig the ML model
```

```python
df.head()
```

```python
km.predict(df[['score','income']])
```

```python
df['new_cluster1']=km.predict(df[['score','income']])
```

```python
df.head()
```

```python
df0= df[df.new_cluster1==0]
```

```python
df0.head()
```

```python
df0.shape
```

```python
#Income and Score plot of df0(cluster 0)
plt.scatter(df0.score,df0.income, color="green")
plt.xlabel("Score in (1 to 100)")
plt.ylabel("Income")
plt.title("Age and score plot")
```

```python
df0= df[df.new_cluster1==0]
df1= df[df.new_cluster1==1]
df2= df[df.new_cluster1==2]
df3= df[df.new_cluster1==3]
df4= df[df.new_cluster1==4]
df5= df[df.new_cluster1==5]
df6= df[df.new_cluster1==6]
df7= df[df.new_cluster1==7]

plt.figure(figsize=(10,8))

#Income and Score plot of df0-df7(cluster 0 to cluster 7)
plt.scatter(df0.score,df0.income, color="green", label="_cluster 0") #at first index i used underscore;So this label is not displayed
plt.scatter(df1.score,df1.income, color="orange", label='cluster 1')
plt.scatter(df2.score,df2.income, color="black", label='cluster 2')
plt.scatter(df3.score,df3.income, color="yellow", label='cluster 3')
plt.scatter(df4.score,df4.income, color="purple", label='cluster 4')
plt.scatter(df5.score,df5.income, color="blue", label='cluster 5')
plt.scatter(df6.score,df6.income, color="pink", label='cluster 6')
plt.scatter(df7.score,df7.income, color="red", label='cluster 7')


plt.xlabel("Score in (1 to 100)")
plt.ylabel("Income")
plt.title("Age and score plot")

plt.legend() #to create the shape of rectangle box
plt.show() #and finally to visualize we have to call show function
```

```python
centre1= km.cluster_centers_
centre1
```

```python
#All the cluster with centroid

df0= df[df.new_cluster1==0]
df1= df[df.new_cluster1==1]
df2= df[df.new_cluster1==2]
df3= df[df.new_cluster1==3]
df4= df[df.new_cluster1==4]
df5= df[df.new_cluster1==5]
df6= df[df.new_cluster1==6]
df7= df[df.new_cluster1==7]

plt.figure(figsize=(10,8))

#Income and Score plot of df0-df7(cluster 0 to cluster 7)
plt.scatter(df0.score,df0.income, color="green", label="_cluster 0") #at first index i used underscore;So this label is not displayed
plt.scatter(df1.score,df1.income, color="orange", label='cluster 1')
plt.scatter(df2.score,df2.income, color="black", label='cluster 2')
plt.scatter(df3.score,df3.income, color="yellow", label='cluster 3')
plt.scatter(df4.score,df4.income, color="purple", label='cluster 4')
plt.scatter(df5.score,df5.income, color="blue", label='cluster 5')
plt.scatter(df6.score,df6.income, color="pink", label='cluster 6')
plt.scatter(df7.score,df7.income, color="red", label='cluster 7')

plt.scatter(centre1[:,0],centre1[:,1], color="red", marker="+")


plt.xlabel("Score in (1 to 100)")
plt.ylabel("Income")
plt.title("Income and score plot")

plt.legend() #to create the shape of rectangle box
plt.show() #and finally to visualize we have to call show function
```

```python
wcss=[]
cluster_range= range(1,15)

for i in cluster_range:
    km2 = KMeans(n_clusters=i)
    km2.fit(df[['score','income']])
    wcss.append(km2.inertia_) #inertia is the value of WCSS
 
```

```python
wcss
```

# VIZ

```python
plt.plot(cluster_range,wcss, marker='o')
plt.ylabel("Value of WCSS")
plt.xlabel("Clusters")
```
